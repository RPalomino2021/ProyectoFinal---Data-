```python
dbutils.widgets.removeAll()
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql import functions as F
```

```python
dbutils.widgets.text("catalogo", "catalog_rp")
dbutils.widgets.text("esquema_source", "silver")
dbutils.widgets.text("esquema_sink", "gold")
```

```python
catalogo = dbutils.widgets.get("catalogo")
esquema_source = dbutils.widgets.get("esquema_source")
esquema_sink = dbutils.widgets.get("esquema_sink")
```

```python
df_automobile_transformed = spark.table(f"{catalogo}.{esquema_source}.automobile_transformed")
```

```python
df_transformed = df_automobile_transformed.groupBy(col("Marca")).agg(
                                                     count(col("TipoVehiculo")).alias("conteo"),
                                                     max(col("Precio")).alias("max_altitude"),
                                                     min(col("Precio")).alias("min_altitude")
                                                     ).orderBy(col("Marca").desc())
```

```python
df_transformed.write.mode("overwrite").saveAsTable(f"{catalogo}.{esquema_sink}.gold_automobile_partitioned")
```
