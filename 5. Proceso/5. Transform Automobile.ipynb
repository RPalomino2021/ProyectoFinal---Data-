```python
dbutils.widgets.removeAll()
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql import functions as F
```

```python
dbutils.widgets.text("catalogo", "catalog_RP")
dbutils.widgets.text("esquema_source", "bronze")
dbutils.widgets.text("esquema_sink", "silver")
```

```python
catalogo = dbutils.widgets.get("catalogo")
esquema_source = dbutils.widgets.get("esquema_source")
esquema_sink = dbutils.widgets.get("esquema_sink")
```

```python
df_automobile = spark.table(f"{catalogo}.{esquema_source}.automobile")
```

```python
from pyspark.sql.functions import col

df_automobile = (
    df_automobile
        .dropna(how="all")  # Elimina filas completamente nulas
        .filter(col("Precio") != "?")  # Elimina registros donde Price sea "?"
)
```

```python
df_automobile.write.mode("overwrite").saveAsTable(f"{catalogo}.{esquema_sink}.automobile_transformed")
```
