```sql
-- =========================================================
-- ELIMINAR SCHEMAS (ORDEN CORRECTO: HIJOS PRIMERO)
-- =========================================================

-- Elimina el esquema gold y todo su contenido (tablas, vistas, funciones)
DROP SCHEMA IF EXISTS catalog_RP.gold CASCADE;

-- Elimina el esquema silver y todo su contenido
DROP SCHEMA IF EXISTS catalog_RP.silver CASCADE;

-- Elimina el esquema bronze y todo su contenido
DROP SCHEMA IF EXISTS catalog_RP.bronze CASCADE;
```

```sql
-- =========================================================
-- ELIMINAR CATÁLOGO
-- =========================================================

-- Elimina el catálogo catalog_dev
-- Solo funcionará si ya no tiene schemas dentro
DROP CATALOG IF EXISTS catalog_RP CASCADE;
```

```sql
-- =========================================================
-- ELIMINAR EXTERNAL LOCATIONS
-- =========================================================

-- Elimina la external location de metastore
DROP EXTERNAL LOCATION IF EXISTS `extl-metastore`;
-- Elimina la external location de gold
DROP EXTERNAL LOCATION IF EXISTS `extl-gold`;

-- Elimina la external location de silver
DROP EXTERNAL LOCATION IF EXISTS `extl-silver`;

-- Elimina la external location de bronze
DROP EXTERNAL LOCATION IF EXISTS `extl-bronze`;
```
