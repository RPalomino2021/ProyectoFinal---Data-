```sql
-- =========================================================
-- CREACIÓN DE GRUPOS
-- =========================================================

-- Grupo para usuarios que solo trabajarán en BRONZE
CREATE GROUP IF NOT EXISTS project_bronze_users;

-- Grupo administrador del proyecto (Reynaldo)
CREATE GROUP IF NOT EXISTS project_admin;
```

```sql
-- =========================================================
-- AGREGAR USUARIOS A LOS GRUPOS
-- =========================================================

-- Usuarios ficticios con acceso limitado a Bronze
ALTER GROUP project_bronze_users ADD USER 'leslyhal@gmail.com';
ALTER GROUP project_bronze_users ADD USER 'repalomino.cibertec@gmail.com';

-- Usuario administrador con control total
ALTER GROUP project_admin ADD USER 'palomino.reynaldo.95@gmail.com';
```

```sql
-- =========================================================
-- PERMISOS A NIVEL DE CATÁLOGO
-- =========================================================

-- Permite a ambos grupos usar el catálogo
GRANT USE CATALOG 
ON CATALOG catalog_RP 
TO `project_bronze_users`;

GRANT USE CATALOG 
ON CATALOG catalog_RP 
TO `project_admin`;
```

```sql
-- =========================================================
-- PERMISOS PARA USUARIOS BRONZE (ACCESO LIMITADO)
-- =========================================================

-- Solo pueden trabajar dentro del schema Bronze
GRANT USE SCHEMA, CREATE, SELECT, MODIFY
ON SCHEMA catalog_RP.bronze
TO `project_bronze_users`;

-- Permiso para leer y escribir archivos en la external location Bronze
GRANT READ FILES, WRITE FILES
ON EXTERNAL LOCATION `extl-bronze`
TO `project_bronze_users`;
```

```sql
-- =========================================================
-- PERMISOS ADMINISTRADOR (REYNALDO - ACCESO TOTAL)
-- =========================================================

-- Permisos completos en todos los schemas
GRANT USE SCHEMA, CREATE, SELECT, MODIFY
ON SCHEMA catalog_RP.bronze
TO `project_admin`;

GRANT USE SCHEMA, CREATE, SELECT, MODIFY
ON SCHEMA catalog_RP.silver
TO `project_admin`;

GRANT USE SCHEMA, CREATE, SELECT, MODIFY
ON SCHEMA catalog_RP.gold
TO `project_admin`;

-- Permisos completos en external locations
GRANT READ FILES, WRITE FILES
ON EXTERNAL LOCATION `extl-bronze`
TO `project_admin`;

GRANT READ FILES, WRITE FILES
ON EXTERNAL LOCATION `extl-silver`
TO `project_admin`;

GRANT READ FILES, WRITE FILES
ON EXTERNAL LOCATION `extl-gold`
TO `project_admin`;

-- Permiso adicional para crear tablas externas
GRANT CREATE EXTERNAL TABLE
ON EXTERNAL LOCATION `extl-bronze`
TO `project_admin`;

GRANT CREATE EXTERNAL TABLE
ON EXTERNAL LOCATION `extl-silver`
TO `project_admin`;

GRANT CREATE EXTERNAL TABLE
ON EXTERNAL LOCATION `extl-gold`
TO `project_admin`;
```
